# Building a Movie Recommendation System using Python


```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import difflib
```

```python
movies=pd.read_csv('movies.csv')
pd.set_option('display.max_columns',None)
movies.head()
```

```python
movies.shape
```

```python
movies.isnull().sum()
```

```python
movies.value_counts().sum()
```

```python
selected_feature=['genres','keywords','popularity','tagline','cast','director']
print(selected_feature)
```

```python
for feature in selected_feature:
    movies[feature]=movies[feature].fillna('')
```

```python
combined_feature=movies['genres']+''+movies['keywords']+''+movies['tagline']+''+''+movies['cast']+''+movies['director']
```

```python
print(combined_feature)
```

```python
vectorizer=TfidfVectorizer()
```

```python
feature_vector=vectorizer.fit_transform(combined_feature)
```

```python
print(feature_vector)
```

```python
similarity=cosine_similarity(feature_vector)
```

```python
movie_name=input('Enter Yours Favorite Movie:   ')
```

```python
movie_title=movies['title'].tolist()
```

```python
find_close_match=difflib.get_close_matches(movie_name,movie_title)
print(find_close_match)
```

```python
close_match=find_close_match[0]
close_match
```

```python
index_movie=movies[movies.title==close_match]['index'].values[0]
index_movie
```

```python
similarity_score=list(enumerate(similarity[index_movie]))

similarity_score
```

```python
sorted_similarity=sorted(similarity_score,key=lambda x:x[1],reverse=True)
```

```python
sorted_similarity
```

```python
print('Movies Recommend For You:   ')

i=1

for movie in sorted_similarity:
    index=movie[0]

    title_movie=movies[movies.index==index]['title'].values[0]

    if(i<21):

      print(i,'.',title_movie)
      i+=1
```

```python
movie_name=input('Enter Yours Favorite Movie:   ')

movie_title=movies['title'].tolist()


find_close_match=difflib.get_close_matches(movie_name,movie_title)

close_match=find_close_match[0]

index_movie=movies[movies.title==close_match]['index'].values[0]

similarity_score=list(enumerate(similarity[index_movie]))

sorted_similarity=sorted(similarity_score,key=lambda x:x[1],reverse=True)

print('Movies Recommend For You:   ')

i=1

for movie in sorted_similarity:
    index=movie[0]

    title_movie=movies[movies.index==index]['title'].values[0]

    if(i<21):

      print(i,'.',title_movie)
      i+=1
```
